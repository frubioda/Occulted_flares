# Some analysis of the RHESSI flare list

## Introduction
We analyze some aspects of the RHESSI flare list, to be found at http://hesperia.gsfc.nasa.gov/hessidata/dbase/hessi_flare_list.txt

More information on the flare list content is at https://www.ssl.berkeley.edu/~jimm/hessi/hsi_flare_list.html

This text file on the web has not all the information stored. We can generate a text file on our own with an apropriate idl routine, "my_write_flarelist.pro" which generates "flarelist_raw.txt" and has about 110000 flares (End 2015).
We read this file with pandas read_table and do the data filtering for the flares that interest us with the python/pandas tools.

For now, in the IDL script the list is already preprocessed to only hold flares that have emission above 12keV and
a valid solar position.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
sns.set_style('white')
matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['figure.figsize'] = (1.4*10., 10.0)
#matplotlib.style.use('ggplot')
#import healpy as hp
```

## Read and reduce data

Reads the data from the flarelist text file into pandas data frame. The start, end and peak time of the flare
are read as date-times. Flares without goes class entry are discarded for now. A 'level' is calculated from the goes class as a quantifier for the flare strength.
The flare peak time, position and active region identifier might be the best things to correlate this database to others.

Other interesting quantities could be:
- peakcnt: peak count rate in corrected counts (probably not so good, since only a countrate, J. McTiernan indicated that this is not a good flare characteristic)
- peakflux: The photon flux in photons/(sec-cm^2) for the peak 4 second interval of the flare, calculated for the energy range greater than 6 keV
- totfluence: The photon fluence in photons/cm^2 for the flare interval above 6 keV
- duration: Duration of the flare in seconds

The GOES levels at the RHESSI peak of the flare in both GOES channels are also included (goes_l1, goes_l2)

```python
# define the names of the columns (could alternatively be read from the header)
colnames = ['idn','tstart','tend','tpeak','ehigh','peakcnt','xpos','ypos','goes',\
            'peakflux','totfluence','duration','goes_l1','goes_l2','active_region']
# read data from file
data = pd.read_table('flarelist_raw.txt', \
                     header=0, skiprows=0, sep=',', \
                     names=colnames, parse_dates=['tstart','tend','tpeak'], engine='python')
# cast to correct types
data['goes'] = data['goes'].astype(str)
data['ehigh'] = data['ehigh'].astype(float)
data['xpos'] = data['xpos'].astype(float)
data['ypos'] = data['ypos'].astype(float)
data['peakcnt'] = data['peakcnt'].astype(int)
# index by tstart and only date
data.index = pd.DatetimeIndex(data.tstart)#.date
# drop empty goes class and add column for goes class equivalent level
data['goes'] = data['goes'].str.strip()
data['goes'].replace('', np.nan, inplace=True)
data.dropna(inplace=True)
data['goes'] = data['goes'].str[0:4]
data.loc[:,'class'] = pd.Series(data['idn'], index=data.index)
data['class'].loc[data['goes'].str.contains('A')] = 1
data['class'].loc[data['goes'].str.contains('B')] = 2
data['class'].loc[data['goes'].str.contains('C')] = 3
data['class'].loc[data['goes'].str.contains('M')] = 4
data['class'].loc[data['goes'].str.contains('X')] = 5
data.loc[:,'level'] = pd.Series(data['idn'], index=data.index)
data['level'].loc[data['goes'].str.contains('A')] = data['goes'].loc[data['goes'].str.contains('A')].str[1:4].astype(float)*1e-8
data['level'].loc[data['goes'].str.contains('B')] = data['goes'].loc[data['goes'].str.contains('B')].str[1:4].astype(float)*1e-7
data['level'].loc[data['goes'].str.contains('C')] = data['goes'].loc[data['goes'].str.contains('C')].str[1:4].astype(float)*1e-6
data['level'].loc[data['goes'].str.contains('M')] = data['goes'].loc[data['goes'].str.contains('M')].str[1:4].astype(float)*1e-5
data['level'].loc[data['goes'].str.contains('X')] = data['goes'].loc[data['goes'].str.contains('X')].str[1:4].astype(float)*1e-4
data['level'] = data['level'].astype(float)

print len(data), 'Flares with goes class in database'
```

```python
# create separate datasets for each goes class
datax = data[data['goes'].str.contains('X')]
datam = data[data['goes'].str.contains('M')]
datab = data[data['goes'].str.contains('B')]
data.columns
```

```python
# plot level distribution
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
limb = plt.Circle((0,0),960,color='black',fill=False)
ax1.set_xlim([-1100,1100])
ax1.set_ylim([-1100,1100])
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.add_artist(limb)
ax1.scatter(datax.loc[:,['xpos']],datax.loc[:,['ypos']],s=datax.loc[:,['level']]*1e7,c='purple',alpha=0.3)
ax1.scatter(datam.loc[:,['xpos']],datam.loc[:,['ypos']],s=datam.loc[:,['level']]*1e7,c='blue',alpha=0.2)
ax1.scatter(datab.loc[:,['xpos']],datab.loc[:,['ypos']],s=datab.loc[:,['level']]*1e7,c='green',alpha=0.5)
plt.show()
```

```python
# plot ehigh distribution
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
limb = plt.Circle((0,0),960,color='black',fill=False)
ax1.set_xlim([-1100,1100])
ax1.set_ylim([-1100,1100])
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.add_artist(limb)
ax1.scatter(datax.loc[:,['xpos']],datax.loc[:,['ypos']],s=datax.loc[:,['ehigh']],c='purple',alpha=0.3)
ax1.scatter(datam.loc[:,['xpos']],datam.loc[:,['ypos']],s=datam.loc[:,['ehigh']],c='blue',alpha=0.3)
ax1.scatter(datab.loc[:,['xpos']],datab.loc[:,['ypos']],s=datab.loc[:,['ehigh']],c='green',alpha=0.3)
plt.show()
```

```python
# plot peakflux distribution
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
limb = plt.Circle((0,0),960,color='black',fill=False)
ax1.set_xlim([-1100,1100])
ax1.set_ylim([-1100,1100])
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.add_artist(limb)
scale = 10
ax1.scatter(datax.loc[:,['xpos']],datax.loc[:,['ypos']],s=scale*np.log(datax.loc[:,['peakflux']]),c='purple',alpha=0.3)
ax1.scatter(datam.loc[:,['xpos']],datam.loc[:,['ypos']],s=scale*np.log(datam.loc[:,['peakflux']]),c='blue',alpha=0.3)
ax1.scatter(datab.loc[:,['xpos']],datab.loc[:,['ypos']],s=scale*np.log(datab.loc[:,['peakflux']]),c='green',alpha=0.3)
plt.show()
```
